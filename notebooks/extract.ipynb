#### Extract and save the data

```python
import pandas as pd
import requests
from time import sleep
from datetime import datetime
import random
from datetime import datetime, timedelta

# === Get previous month range ===
today = datetime.today()
first_of_this_month = datetime(today.year, today.month, 1)
last_of_prev_month = first_of_this_month - timedelta(days=1)
first_of_prev_month = datetime(last_of_prev_month.year, last_of_prev_month.month, 1)

start_date = first_of_prev_month.strftime("%Y-%m-%d")
end_date = last_of_prev_month.strftime("%Y-%m-%d")
month_str = first_of_prev_month.strftime("%Y_%m")

print(f"📅 Collecting weather data for: {start_date} to {end_date}")

# Your full districts DataFrame is assumed to be defined already
# Function remains the same
def fetch_weather_with_retry(latitude, longitude, start_date, end_date, max_retries=5):
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": start_date,
        "end_date": end_date,
        "daily": "temperature_2m_max,temperature_2m_min,precipitation_sum,relative_humidity_2m_mean,sunshine_duration",
        "timezone": "Africa/Kampala"
    }
    backoff = 5
    for attempt in range(1, max_retries + 1):
        try:
            response = requests.get(url, params=params)
            if response.status_code == 429:
                wait_time = int(response.headers.get("Retry-After", backoff))
                print(f"Rate limited. Waiting {wait_time} seconds...")
                sleep(wait_time)
                backoff *= 2
                continue
            response.raise_for_status()
            return response.json()
        except Exception as e:
            print(f"Attempt {attempt} failed: {str(e)}")
            sleep(backoff)
            backoff *= 2
    print(f"Max retries exceeded.")
    return None

# Set date range for prev month
# Get today's date
today = datetime.today()

# First day of the current month
first_day_current_month = datetime(today.year, today.month, 1)

# Last day of the previous month
last_day_prev_month = first_day_current_month - timedelta(days=1)

# First day of the previous month
first_day_prev_month = datetime(last_day_prev_month.year, last_day_prev_month.month, 1)

# Format as strings
start_date = start_date
end_date = end_date

# Store results
weather_data = []

# Loop through each district and fetch data
for idx, row in districts.iterrows():
    print(f"Fetching data for {row['district']}...")
    result = fetch_weather_with_retry(row['latitude'], row['longitude'], start_date, end_date)

    if result and 'daily' in result:
        daily_data = result['daily']
        for i in range(len(daily_data['time'])):
            weather_data.append({
                "district": row['district'],
                "date": daily_data['time'][i],
                "temp_max": daily_data.get("temperature_2m_max", [None]*len(daily_data['time']))[i],
                "temp_min": daily_data.get("temperature_2m_min", [None]*len(daily_data['time']))[i],
                "precipitation": daily_data.get("precipitation_sum", [None]*len(daily_data['time']))[i],
                "humidity": daily_data.get("relative_humidity_2m_mean", [None]*len(daily_data['time']))[i],
                "sunshine": daily_data.get("sunshine_duration", [None]*len(daily_data['time']))[i]
            })

    sleep(random.uniform(0.5, 1.5))  # polite delay

# Convert to DataFrame
df_weather = pd.DataFrame(weather_data)
df_weather["month"] = month_str

df_weather.to_csv(f"raw_weather_data_{month_str}.csv", index=False)

# Preview
print(df_weather.head())
```
